QUESTION TWO: Algorithm Analysis and Complexity

a) Define the following asymptotic notations: (3 Marks)

 i. Big-O notation (O) specifies the upper bound of the time or space complexity of an algorithm. A function f(n) is O(g(n)) if there exist constants c > 0 and n₀ such that f(n) ≤ c g(n) for all n ≥ n₀. With a focus on the algorithm's performance as n rises, it presents the worst-case scenario. 
 
 ii. Big-Theta notation defines a narrow border for the complexity of an algorithm, spanning both an upper and lower limit. The function f(n) is Θ(g(n)) if there are constants c₁ > 0, c₂ > 0, and n₀ so c₁ g(n) ≤ f(n) ≤ c₂ g(n) for all n ≥ n₀. It shows that the rate of growth is exactly of g(n). 
 
 iii. Big-Omega notation characterizes the lower limit of an algorithm's complexity. The function f(n) is said to be Ω(g(n)) if constants c > 0 and n0 exist such that f(n) ≥ c g(n) for all n ≥ n0. It denotes the lowest possible growth rate or the most desirable result.

QUESTION THREE: Searching and Sorting Algorithms



```python
#a) Write a Python function to implement linear search to find the value 32. (2 Marks)

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i # Return the index if found
    return -1 # Return -1 if not found


arr = [21, 10, 45, 32, 67, 5, 18]    
index = linear_search(arr, 32)
print(f"Value 32 is found at index: {index}")
```

```python
#b) Write a Python function to implement binary search to find the value 45.

def binary_search(arr, target):
    arr.sort()  # Sort the list first as per assumption
    left, right = 0, len(arr) -1
    while left < right:
        mid = (left + right) //2
        if arr[mid] == target:
            return mid # Return the index if found
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1 # Return -1 if not found

arr = [21, 10, 45, 32, 67, 5, 18]     
index = binary_search(arr, 45)
print(f"Value 45 is found at index: {index} (in sorted list)")

```

```python
#c) Implement bubble sort to sort the list in ascending order. (2 Marks)

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]: #Swap if the element found is greater than the next element
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [21, 10, 45, 32, 67, 5, 18]            
sorted_arr = bubble_sort(arr)
print(f"Sorted list: {sorted_arr}")
```

QUESTION FOUR: Linked Lists Implementation


```python
#a) Define a Node class in Python for a singly linked list. (2 Marks)

class node:
    def __init__(self, data):
        self.data = data # Store the data
        self.next = None  # Pointer to the next node

```

```python
#b) Write a LinkedList class with the following operations: (4 Marks)

class LinkedList:
    def __init__(self):
        self.head = None

    #i. Insert a node at the beginning
    def insert_at_beginning(self, data):     
        new_node = node(data)
        new_node.next = self.head
        self.head = new_node  # Initialize the head of the list


    #ii. Insert a node at the end
    def insert_at_end(self, data):
        new_node = node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    #iii. Display all elements in the linked list
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
            

```

a) Queue. Follows the First in First out(FIFO) principle, where the first student to arrive is first to be served. 

```python
#b) Write a Python program that simulates this process using the chosen data structure with the following operations: (4 Marks)

class Queue:
    def __init__(self):
        self.items = []

    #i. Enqueue (add a customer to the queue)
    def enqueue(self, item):
        self.items.append(item)

    #ii. Dequeue (serve a customer and remove them from the queue)
    def dequeue(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return "Queue is empty"

    #iii. Display the current queue
    def display(self):
        print("Current Queue:", self.items)

    def is_empty(self):
        return len(self.items) == 0 
```
